```python
import matplotlib.pyplot as plt

# Sample data (e.g., churn rate over time)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
churn_rate = [4.8, 4.85, 4.9, 4.95, 5.0, 5.1]
```

```python
# Proper scaling
plt.figure()
plt.plot(months, churn_rate)
plt.ylim(0, 6)  # Proper y-axis starting at zero
plt.xlabel("Month")
plt.ylabel("Churn Rate (%)")
plt.title("Properly Scaled Line Chart")
plt.show()
```

```python
# Improper scaling
plt.figure()
plt.plot(months, churn_rate)
plt.ylim(4.7, 5.2)  # Truncated y-axis exaggerates variation
plt.xlabel("Month")
plt.ylabel("Churn Rate (%)")
plt.title("Improperly Scaled Line Chart (Misleading)")
plt.show()
```
