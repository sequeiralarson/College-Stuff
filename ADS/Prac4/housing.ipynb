# Boston Data
A data set containing housing values in 506 suburbs of Boston.

-crim: per capita crime rate by town.

-zn: proportion of residential land zoned for lots over 25,000 sq.ft.

-indus: proportion of non-retail business acres per town.S

-chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

-nox: nitrogen oxides concentration (parts per 10 million).

-rm: average number of rooms per dwelling.

-age: proportion of owner-occupied units built prior to 1940.

-dis: weighted mean of distances to five Boston employment centres.

-rad: index of accessibility to radial highways.

-tax: full-value property-tax rate per $10,000.

-ptratio: pupil-teacher ratio by town.

-lstat: lower status of the population (percent).

-medv: median value of owner-occupied homes in $1000s.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np 

sns.set(style="whitegrid")

# Load data
df = pd.read_csv("HousingData.csv")

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle("Distribution of House Prices", fontsize=16)

sns.histplot(df["MEDV"], bins=30, kde=True, ax=axes[0])
axes[0].set_title("Histogram of MEDV")

sns.boxplot(x=df["MEDV"], ax=axes[1])
axes[1].set_title("Boxplot of MEDV")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("MEDV vs Key Predictors", fontsize=16)

sns.scatterplot(x="RM", y="MEDV", data=df, ax=axes[0, 0])
axes[0, 0].set_title("MEDV vs Rooms (RM)")

sns.scatterplot(x="LSTAT", y="MEDV", data=df, ax=axes[0, 1])
axes[0, 1].set_title("MEDV vs Lower Status (LSTAT)")

sns.scatterplot(x="CRIM", y="MEDV", data=df, ax=axes[1, 0])
axes[1, 0].set_title("MEDV vs Crime Rate (CRIM)")

sns.scatterplot(x="NOX", y="MEDV", data=df, ax=axes[1, 1])
axes[1, 1].set_title("MEDV vs Pollution (NOX)")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Location & Policy Effects on House Prices", fontsize=16)

sns.boxplot(x="CHAS", y="MEDV", data=df, ax=axes[0, 0])
axes[0, 0].set_title("MEDV by Charles River Proximity")

sns.scatterplot(x="DIS", y="MEDV", data=df, ax=axes[0, 1])
axes[0, 1].set_title("MEDV vs Distance to Employment Centers")

sns.scatterplot(x="TAX", y="MEDV", data=df, ax=axes[1, 0])
axes[1, 0].set_title("MEDV vs Property Tax Rate")

sns.scatterplot(x="PTRATIO", y="MEDV", data=df, ax=axes[1, 1])
axes[1, 1].set_title("MEDV vs Pupil–Teacher Ratio")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

corr = df.corr()
sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0,
    linewidths=0.5
)

plt.title("Correlation Matrix")
plt.show()
```
