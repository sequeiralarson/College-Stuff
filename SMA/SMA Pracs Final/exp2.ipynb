```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = r"https://quotes.toscrape.com"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    print(soup.prettify()[:1000])
else:
    print(f"Failed to get response. \nResponse: {response}")
```

```python
quotes, authors, tags = [], [], []

quote_container = soup.find_all('div', class_='quote')

for container in quote_container:
    quote_text = container.find('span', class_='text').text.strip()
    quotes.append(quote_text)

    author_text = container.find('small', class_='author').text.strip()
    authors.append(author_text)

    tag_list = container.find('div', class_='tags').find_all('a', class_='tag')
    tags_list = [tag.text for tag in tag_list]
    tags.append(tags_list)

data = {
    "tags": tags,
    "quotes": quotes,
    "author": authors,
}

df = pd.DataFrame(data)
df.to_csv("quotes", index=False)
```
