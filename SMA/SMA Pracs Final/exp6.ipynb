```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a random graph (Erdős–Rényi model)
n = 30  # Number of nodes
p = 0.2  # Probability of edge creation
G = nx.erdos_renyi_graph(n, p)

# Visualize the graph
plt.figure(figsize=(8, 8))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, font_size=12)
plt.title('Random Social Network (Erdős–Rényi Model)')
plt.show()
```

```python
degrees = [G.degree(n) for n in G.nodes()]
plt.hist(degrees, bins=range(min(degrees), max(degrees)+1), alpha=0.75, color='green')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:", degree_centrality)
```

```python
betweenness_centrality = nx.betweenness_centrality(G)
print("Betweenness Centrality:", betweenness_centrality)
```

```python
closeness_centrality = nx.closeness_centrality(G)
print("Closeness Centrality:", closeness_centrality)
```

```python
import community.community_louvain as community_louvain  # This is correct after installing python-louvain

# Example Graph (You can replace this with your own graph)
G = nx.karate_club_graph()  # You can use any graph (e.g., your own loaded network)

# Detect communities using Louvain method
partition = community_louvain.best_partition(G)

print("Communities detected:", partition)

# Visualize the communities
pos = nx.spring_layout(G)  # Positioning nodes using spring layout

plt.figure(figsize=(8, 8))
# Draw nodes with community color
nx.draw_networkx_nodes(G, pos, node_color=list(partition.values()), cmap='rainbow', node_size=500)
# Draw edges
nx.draw_networkx_edges(G, pos, alpha=0.7)
# Draw labels
nx.draw_networkx_labels(G, pos, font_size=12)

plt.title('Community Detection in Network')
plt.show()
```

```python
clustering_coeff = nx.clustering(G)
print("Clustering Coefficients:", clustering_coeff)
```

```python
diameter = nx.diameter(G)
print("Network Diameter:", diameter)
```

```python
# Identify top 5 most influential nodes by betweenness centrality
top_influencers = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 Influencers:", top_influencers)
```
