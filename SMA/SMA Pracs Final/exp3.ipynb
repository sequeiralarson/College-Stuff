```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'name': ['Alice', 'Bob', None, 'David'],
    'age': [25, None, 23, 22]
})

# Remove rows with missing values
df_cleaned = df.dropna()

# Fill missing values with a specific value (e.g., 0 for 'age')
df_filled = df.fillna({'age': 0})

print(df_cleaned)
print(df_filled)
```

```python
# Example DataFrame with duplicates
df = pd.DataFrame({
    'post': ['Post 1', 'Post 2', 'Post 1'],
    'likes': [150, 200, 150]
})

# Remove duplicate rows based on all columns
df_no_duplicates = df.drop_duplicates()

# Remove duplicates based on a specific column (e.g., 'post')
df_no_duplicates_by_post = df.drop_duplicates(subset=['post'])

print(df_no_duplicates)
print(df_no_duplicates_by_post)
```

```python
# Convert 'likes' column to integer type
df['likes'] = df['likes'].astype(int)
```

```python
# Example DataFrame with categorical data
df = pd.DataFrame({
    'platform': ['Facebook', 'Twitter', 'Instagram']
})

# Convert 'platform' column to one-hot encoding
df_encoded = pd.get_dummies(df, columns=['platform'])

print(df_encoded)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Example numerical data
data = [[1000], [2000], [3000], [4000]]

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(data)

print(scaled_data)
```

```python
from sklearn.preprocessing import StandardScaler

# Example numerical data
data = [[1000], [2000], [3000], [4000]]

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(data)

print(scaled_data)
```

```python
from nltk.tokenize import word_tokenize

text = "Hello, this is an example post. #Example"
tokens = word_tokenize(text.lower())  # Tokenize and convert to lowercase

print(tokens)
```

```python
from nltk.corpus import stopwords

# Load the stopwords list
stop_words = set(stopwords.words('english'))

# Example text
tokens = ["this", "is", "an", "example", "post"]

# Filter out stop words
filtered_tokens = [word for word in tokens if word not in stop_words]

print(filtered_tokens)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

# Example words
words = ['running', 'runner', 'runs', 'ran']

# Apply stemming
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

```python
# Example DataFrame with numerical data
df = pd.DataFrame({
    'likes': [100, 200, 1000, 5000, 100000]
})

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['likes'].quantile(0.25)
Q3 = df['likes'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define outliers
outlier_condition = (df['likes'] < (Q1 - 1.5 * IQR)) | (df['likes'] > (Q3 + 1.5 * IQR))

# Remove outliers
df_no_outliers = df[~outlier_condition]

print(df_no_outliers)
```
